# Project Assumptions and Analytical Criteria 

**Purpose**

Certain analytical criteria and thresholds were not explicitly defined in the project statement. Therefore, reasonable, data-driven assumptions were established to ensure consistency throughout the analysis. These criteria were selected using common statistical practices and are summarized below.

# Task 2 – Descriptive Transactional Analysis

Dormant Account: An account was considered dormant if the gap between two consecutive transactions was 60 days (approximately two months) or more, as specified in the project requirement.

# Task 3 – Customer Profile Building

**Activity Level Classification**

**Accounts were categorized based on transaction frequency:**

- High Activity: 5 or more transactions
- Medium Activity: 3–4 transactions
- Low Activity: 1–2 transactions

**Balance & Transaction Volume Segmentation:** 

- The median of Average Account Balance was used to classify accounts into High Balance and Low Balance groups.
- The median of Total Transaction Volume was used to classify accounts into High Volume and Low Volume groups.

The median was selected because it provides a balanced segmentation while minimizing the influence of extreme values.

**High-Net Inflow Accounts:**

- Accounts with Net Inflow values greater than the 75th percentile were classified as High-Net Inflow Accounts.

**Near-Zero Balance Accounts:** 

- Accounts with an Average Balance of ₹1,000 or less were classified as Near-Zero Balance Accounts for analytical purposes.

# Task 4 – Financial Risk Identification

**Large Withdrawal:**

A transaction was classified as a Large Withdrawal if its amount was greater than or equal to the 75th percentile of all withdrawal transactions.

**Frequent Large Withdrawals:** 

Accounts with two or more Large Withdrawals were classified as High-Risk Withdrawal Accounts.

**Frequent Overdrafts:** 

Accounts with two or more overdraft occurrences were classified as Frequent Overdraft Accounts.

**Balance Volatility:**

- Balance volatility was measured using the Coefficient of Variation (CV).
- Accounts with a CV greater than or equal to the 75th percentile were classified as High Volatility Accounts.

**Suspicious Customer Criteria:**
Customers satisfying all of the following conditions were classified as suspicious:

- Risk Score > 0.70
- Credit Rating < 450
- Negative Account Balance

These criteria were chosen to identify customers with elevated financial risk.

# Task 6 – Hypothesis Testing

**Customer Segmentation:**

- Accounts were divided into High-Volume and Low-Volume groups using the median Transaction Volume.
- The median was selected to create balanced comparison groups and reduce the influence of extreme values.

**Statistical Significance:**
All hypothesis tests were conducted using a **5% significance level (α = 0.05)**, which is the standard threshold used in statistical analysis.

**Note: All analytical criteria that I used were selected using standard data analytics and statistical practices where explicit criteria were not provided for the project. 
The purpose of these assumptions was to maintain consistency, objectivity, and reproducibility throughout the analysis.**
