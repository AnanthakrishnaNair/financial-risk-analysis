# Task 1: Data Cleaning and Formatting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from fpdf import FPDF
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/goldman_sachs.csv")
```

```python
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.isnull().sum())
df.describe(include="all")
```

```python
categorical_cols = [
    "AccountType",
    "TransactionType",
    "Product",
    "Firm",
    "Region",
    "Manager"
]

for col in categorical_cols:
    print(f"\n===== {col} =====")
    print(df[col].unique())
```

```python
# Converting TransactionDate to datetime format
df["TransactionDate"] = pd.to_datetime(
    df["TransactionDate"],
    format="%d-%m-%Y",
    errors="coerce"
)

# Check whether any dates failed to convert
print("Invalid dates:", df["TransactionDate"].isna().sum())
```

```python
print(df.dtypes)
```

# Observations

- The `TransactionDate` column was originally stored as a string (`object`).
- It was successfully converted into the datetime format.
- No invalid or missing dates were encountered during conversion (`Invalid dates = 0`).
- This conversion enables time-based analysis such as monthly trends, yearly summaries, and dormancy detection in later tasks.

```python
# Standardizing the text columns
str_cols = [
    "AccountType",
    "TransactionType",
    "Product",
    "Firm",
    "Region",
    "Manager",
    "CustomerID",
    "AccountID"
]

for col in str_cols:
    df[col] = df[col].str.strip().str.title()

print("Text columns standardized successfully.")
```

```python
# Find transactions with negative amounts
negative_txns = df[df["TransactionAmount"] < 0]

print("Number of negative transactions:", len(negative_txns))
negative_txns.head(10)
```

```python
negative_txns["TransactionType"].value_counts()
```

```python
# Count negative transaction amounts before cleaning
neg_count = (df["TransactionAmount"] < 0).sum()
print(f"Negative TransactionAmount values before cleaning: {neg_count}")

# Convert all transaction amounts to absolute values
df["TransactionAmount"] = df["TransactionAmount"].abs()

# Verify that no negative values remain
remaining_neg = (df["TransactionAmount"] < 0).sum()
print(f"Negative TransactionAmount values after cleaning: {remaining_neg}")
```

# Observations

During data validation, 24 negative values were identified in the `TransactionAmount` column.

Further investigation showed that negative values were present across multiple transaction types, including **Withdrawal, Payment, 
Transfer, and Deposit**. Since the transaction direction is already explicitly represented by the `TransactionType` column, storing 
negative values in `TransactionAmount` would duplicate the representation of transaction direction and complicate later aggregation.

Therefore, all transaction amounts were converted to their absolute values, while the transaction direction continues to be determined 
by the `TransactionType` column. This ensures consistent and accurate calculations in subsequent analyses. 

```python
# Checking for the RiskScore statistics
print(df["RiskScore"].describe())

# Find values outside the valid range (0 to 1)
invalid_risk = df[
    (df["RiskScore"] < 0) |
    (df["RiskScore"] > 1)
]

print(f"Number of invalid RiskScore values: {len(invalid_risk)}")
```

```python
# Counting invalid RiskScore values
invalid_risk = ((df["RiskScore"] < 0) | (df["RiskScore"] > 1)).sum()
print(f"Invalid RiskScore values before cleaning: {invalid_risk}")

# Clipping the values to a valid range [0, 1]
df["RiskScore"] = df["RiskScore"].clip(lower=0, upper=1)

# Verifying the cleaning
remaining_invalid = ((df["RiskScore"] < 0) | (df["RiskScore"] > 1)).sum()
print(f"Invalid RiskScore values after cleaning: {remaining_invalid}")
```

# Observation

The RiskScore column is expected to contain normalized values between 0 and 1. During validation, 35 values were found outside 
this range.

Since these values are outside the valid domain of a normalized risk score, they were clipped to the nearest valid boundary (0 or 1). 
This approach preserves all records while ensuring that the data remains consistent and suitable for further analysis.

```python
# Display CreditRating 
print(df["CreditRating"].describe())

# Finding invalid CreditRating values
invalid_credit = df[
    (df["CreditRating"] < 300) |
    (df["CreditRating"] > 850)
]

print(f"Invalid CreditRating values: {len(invalid_credit)}")
```

```python
# Creating an overdraft flag
df["IsOverdraft"] = df["AccountBalance"] < 0

# Displaying count of overdraft accounts
print(df["IsOverdraft"].value_counts())
```

# Observation

A new column named `IsOverdraft` was created.

- True indicates that the account balance is negative (overdraft).
- False indicates that the account balance is zero or positive.

This feature will be used in the financial risk analysis performed in later tasks.


```python
# Saving the cleaned dataset
df.to_csv("goldman_sachs_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```
